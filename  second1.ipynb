 # Covariance and Correlation 

  -- Covariance and Correlation are used in statistics and data science to understand the relationship between two variables.

-- covariance = Covariance is a statistical measure that shows how two variables change together.

-- In simple words, it tells us whether two variables increase or decrease together.its renge + infniti to - infiniti

-- Correlation =  is a statistical measure that shows the strength and direction of the relationship between two variables.

-- In simple words, correlation tells us how stron it range is -1 to + 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r"C:\Users\imrat\OneDrive\Desktop\data_science_all_uses_items\Data_set\tips.csv")
dataset.head(3)
```

```python
data_cov = dataset.select_dtypes(["float64","int64"]).cov()
data_cov
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(data_cov,annot=True)
plt.show()
```

# central limit theorem

-- We use the Central Limit Theorem when we do not know the population distribution.
    
-- The Central Limit Theorem is used when we want to analyze population parameters
-- using sample data, especially when the sample size is large (n ≥ 30), because the
-- sampling distribution of the mean becomes approximately normal.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pop_data = [np.random.randint(10,100) for i in range(10000)]
len(pop_data)
pop_table = pd.DataFrame({"pop_data":pop_data})
pop_table.head()
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="pop_data",data=pop_table)
plt.show()
```

```python
sample_mean = []
for no_sample in range(50):
    sample_data = []
    for data in range(500):
        sample_data.append(np.random.choice(pop_data))
    sample_mean.append(np.mean(sample_data))
```

```python
sample_M = pd.DataFrame({"sample_mean":sample_mean})
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="sample_mean",data=sample_M)
plt.show()
```

```python
np.mean(pop_data) ,np.mean(sample_M)
```

```python
np.median(sample_M)
```
