-- Q . a teacher claims that the mean score of students in his class is grater then 82 with a standard deciation of 20 . if a sample of 81 students was selected with a mean score of 90.

```python
import scipy.stats as st
import numpy as np
```

```python
st.norm.ppf(0.95)
```

```python
s_x = 90
pu = 82
pstd =20
n = 81
ap = 0.05
```

```python
z_cal = (s_x - pu) / (pstd /np.sqrt(n))
z_cal
```

```python
z_table = st.norm.ppf(1-ap)
z_table
```

```python
if z_table < z_cal:
    print("ha is right :")
else:
    print("h0 is right :")
```
