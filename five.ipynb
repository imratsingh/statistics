Q.2

-- A study was conducted to investigate whether there is a relationship between 
gender and the preferred genre of music. A sample of 235 people was selected,
and the data collected is shown below. Test at a 5% level of significance
whether there is a significant association between gender and music preference.

Gender	Pop	    HipHop	   Classical	Rock

Male	40	     45	        25 	         10

Female	35	     30	        20	         30

```python
import numpy as np
```

```python
row1 = np.array([40,45,25,10])
row2 = np.array([35,30,20,30])
```

```python
sum_r1 = np.sum(row1)
sum_r2 = np.sum(row2)
sum_row = np.array([sum_r1,sum_r2])
sum_row
```

```python
sum_cal = row1 + row2
sum_cal
```

```python
exp = []
for i in sum_row:
    for j in sum_cal:
        value = (i*j)/235
        exp.append(value)
```

```python
obj = np.array([40,45,25,10,35,30,20,30])
obj
```

```python
np.sum((np.square(obj-exp))/exp)
```
